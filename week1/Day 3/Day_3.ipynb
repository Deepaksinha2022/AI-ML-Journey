```python
## Hour 7

# Remove duplicates from sorted array

lst = [1,1,2,2,3]

def remove_dup(lst):
    i = 0
    for j in range(1,len(lst)):
        if lst[i]!= lst[j]:
            i+=1
            lst[i] = lst[j]
    return lst
print(remove_dup(lst))
```
