```python
def even_odd(n):
    if n%2 ==0:
        return "even"
    else:
        return "odd"
    
print(even_odd(24))
```

```python
## Factorial of a number

def fact(n):
    fct=1
    for i in range(1,n+1):
        fct = i*fct
    return fct

print(fact(5))
```

```python
## Reverse a number
def rev_num(n):
    rev = 0
    while n>0:
        mod = n%10
        rev = rev*10 + mod
        n= n//10
    return rev

print(rev_num(123))
```

```python
## Sum of digits

def sum_digits(n):
    sum = 0
    while n>0:
        mod = n%10
        sum += mod
        n=n//10
    return sum

print(sum_digits(123))
```

```python
## Count of Digit

def count_digit(n):
    count = 0
    while n>0:
        n = n//10
        count+= 1
    return count
print(count_digit(1238963))
```

```python
## Count number of Vowels

def count_vowels(str):
    vow = ["a","e","i","o","u"]
    count = 0
    for i in str:
        if i in vow:
            count+=1
    return count

print(count_vowels("Anvi Pritha"))
```

```python
## Count number of Consonants

def count_vowels(str):
    vow = ["a","e","i","o","u"]
    count = 0
    for i in str:
        if i not in vow:
            count+=1
    return count

print(count_vowels("Jyoti"))
```

```python
## Count of each vowel

def count_vow(s):
    dct = {}
    vowels = ["a","e","i","o","u"]
    
    for ch in s:
        if ch in vowels:
            if ch in dct:
                dct[ch] += 1
            else:
                dct[ch] = 1
                
    return dct

print(count_vow("hellouiiuo"))
```

```python
## Check prime or not

def is_prime(n):
    if n ==0 or n == 1:
        return "invalid"
    if n == 2:
        return "prime"
    for i in range(2,n+1):
        if n%i == 0:
            return "not prime"
        else:
            return "Prime"
        
print(is_prime(79))
```

```python
## Count of Divisor

def divisor_count(num):
    count = 0
    for i in range(1, num+1):
        if num % i == 0:
            count+= 1
    return count

print(divisor_count(17))
        
```

```python
## Sum of All Divisors
def divisor_sum(num):
    sum = 0
    for i in range(1, num+1):
        if num % i == 0:
            sum+= i
    return sum

print(divisor_sum(10))
```
